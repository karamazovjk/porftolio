```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
        df = pd.read_csv("/kaggle/input/datasets/amar5693/screen-time-sleep-and-stress-analysis-dataset/Smartphone_Usage_Productivity_Dataset_50000.csv")
print(df.head())
print(df.info())
print(df.describe())
media_sono = df["Sleep_Hours"].mean()

print(df.shape)
df["Sleep_Hours"].mean()
df["Stress_Level"].mean()


import sqlite3

conn = sqlite3.connect(":memory:")
df.to_sql("smartphone_data", conn, index=False, if_exists="replace")
query = """
SELECT AVG(Sleep_Hours) AS media_sono,
       AVG(Stress_Level) AS media_estresse
FROM smartphone_data
"""
resultado = pd.read_sql(query, conn)
print(resultado)
```
